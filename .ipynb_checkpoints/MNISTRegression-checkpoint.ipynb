```python
import numpy as np
import torch
import torchvision

torch.cuda.is_available()
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(torch.cuda.get_device_name(0))

import data, models, train, measurements
import optimizer as opt
```

```python
train_data = data.load_reg_mnist()
test_data = data.load_reg_mnist(test=True)

model = models.fully_connected_net(28 * 28, 1, [400, 400], 'relu').cuda()
loss_fn = torch.nn.MSELoss()
torch.save({'model_state_dict': model.state_dict()}, './temp/model_init')

output = model(train_data[0].cuda())
print(output)
print(train_data[1].cuda())
print(loss_fn(output, train_data[1].cuda()))
```

```python
lr = 0.1
mom = 0.
batch_size = 5000
phys_bs = 5000
num_epochs = 500

msrs = measurements.Measurement(verbose=True)
msrs.add_train_recorder('MSE', phys_bs, verbose=True)
msrs.add_train_recorder('MSE Sharpness', phys_bs, verbose=True)
msrs.add_train_recorder('Tangent Feature Change', phys_bs, every=10, verbose=False)
msrs.add_train_recorder('Hessian Second Order Term Norm', phys_bs, verbose=True)
msrs.add_train_recorder('MSE Eigenvalues', phys_bs, every=10, verbose=False)
msrs.add_test_recorder('MSE', phys_bs, verbose=True)

optimizer = torch.optim.SGD(model.parameters(), lr=lr, momentum=mom)
```

```python
model.load_state_dict(torch.load('./temp/model_init')['model_state_dict'])
train.train(train_data, test_data, model, loss_fn, optimizer, batch_size, num_epochs, msrs)
```

```python
from matplotlib import pyplot as plt

for rec in msrs.get_train_recorder():
    if rec.get_name() == 'MSE':
        MSE_hist = rec.get_records()
    if rec.get_name() == 'MSE Sharpness':
        MSE_sharp = rec.get_records()
    if rec.get_name() == 'Hessian Second Order Term Norm':
        h_hat_sharp = rec.get_records()
    if rec.get_name() == 'Tangent Feature Change':
        tf_change = rec.get_records()
    if rec.get_name() == 'MSE Eigenvalues':
        top_eigenvalues = rec.get_records()
```

```python
from matplotlib import lines

x = [x for x, y in MSE_hist]
MSEs = [y for x, y in MSE_hist]
sharps = [np.abs(y) for x, y in MSE_sharp]
h_hats = [np.abs(y) for x, y in h_hat_sharp]
change = [y for x, y in tf_change]

_, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax2.plot(x, MSEs, label='MSE', color='yellow')
ax2.plot([x for x, y in tf_change], change, label='Feature Change', color='red')
ax1.plot(x, sharps, label='Hessian Norm', color='blue')
ax1.plot(x, h_hats, label='h_hat Norm', color='green')
ax2.set_yscale('log')
ax1.set_yscale('log')
ax1.set_ylabel('Hessian Norms')
ax2.set_ylabel('MSE/Feature Change')
plt.xlabel('Epoch IDX')
handles, _ = ax1.get_legend_handles_labels()
mse_legend_line = lines.Line2D([0], [0], label='MSE', color='yellow')
tf_legend_line = lines.Line2D([0], [0], label='Feature Change', color='red')
handles.extend([mse_legend_line, tf_legend_line])
ax1.legend(handles=handles)
plt.savefig('HessianNormCmp.png', dpi=400)
```

```python
evecs_hist = [y[1] for x, y in top_eigenvalues]

def evec_diff(evec1, evec2):
    diff_norms = [np.linalg.norm(ev1.cpu().numpy() - ev2.cpu().numpy()) ** 2 for ev1, ev2 in zip(evec1, evec2)]
    evec1_norms = [np.linalg.norm(ev1.cpu().numpy()) ** 2 for ev1 in evec1]
    evec2_norms = [np.linalg.norm(ev2.cpu().numpy()) ** 2 for ev2 in evec2]
    return np.sqrt(2 - np.abs(2 - np.sum(diff_norms)))


evec_changes = [[evec_diff(evec1, evec2) for evec1, evec2 in zip(all_evec1, all_evec2)] for all_evec1, all_evec2 in zip(evecs_hist[:-1], evecs_hist[1:])]
```

```python
sns.heatmap(evec_changes)
plt.xlabel('Eigenvector IDX')
plt.ylabel('Snapshot IDX')
plt.savefig('EigenvecChange.png', dpi=400)
```

```python
evec_mat_hist = []
for snapshot in evecs_hist:
    fl_evecs_list = []
    for evec in snapshot:
        fl_evecs_list.append(np.concatenate([w.cpu().numpy().flatten() for w in evec]))
    snap_mat_form = np.stack(fl_evecs_list)
    evec_mat_hist.append(snap_mat_form)
```

```python
def space_overlap(evecs1, evecs2, top=1):
    assert(top <= evecs1.shape[0])
    m1 = evecs1[:top]
    m2 = evecs2[:top]
    norm = np.linalg.norm(m1 @ m2.T)
    return norm / np.sqrt(top)
```

```python
overlaps = np.zeros((len(evec_mat_hist)-1, 50))
for idx, pair in enumerate(zip(evec_mat_hist[:-1], evec_mat_hist[1:])):
    evec1, evec2 = pair
    for k in range(50):
        num_evecs = k+1
        overlaps[idx, k] = space_overlap(evec1, evec2, top=num_evecs)

print(overlaps)
```

```python
sns.heatmap(overlaps)
plt.xlabel('TopK')
plt.ylabel('Snapshot IDX')
plt.savefig('SubspaceOverlap.png', dpi=400)
```
