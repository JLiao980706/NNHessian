```python
import numpy as np
import torch
import torchvision

torch.cuda.is_available()
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(torch.cuda.get_device_name(0))

import data, models, train, measurements
import optimizer as opt
```

```python
train_data = data.load_mnist(num_per_class=500)
test_data = data.load_mnist(test=True)

model = models.fully_connected_net(28 * 28, 10, [400, 400], 'relu').cuda()
loss_fn = torch.nn.CrossEntropyLoss()
torch.save({'model_state_dict': model.state_dict()}, './temp/model_init')

output = model(train_data[0].cuda())
print(output)
print(train_data[1].cuda())
print(loss_fn(output, train_data[1].cuda()))
```

```python
lr = 0.2
mom = 0.
batch_size = 5000
phys_bs = 5000
num_epochs = 100

msrs = measurements.Measurement(verbose=True)
msrs.add_train_recorder('Cross Entropy', phys_bs, verbose=True)
msrs.add_train_recorder('Accuracy', phys_bs, verbose=True)
msrs.add_train_recorder('Cross Entropy Sharpness', phys_bs, verbose=True)
# msrs.add_train_recorder('Cross Entropy Sharpness V2', phys_bs, verbose=True)
msrs.add_test_recorder('Cross Entropy', phys_bs, verbose=True)
msrs.add_test_recorder('Accuracy', phys_bs, verbose=True)

# optimizer = torch.optim.SGD(model.parameters(), lr=lr, momentum=mom)
optimizer = opt.Adsgd(model.parameters(), lr=lr)
```

```python
model.load_state_dict(torch.load('./temp/model_init')['model_state_dict'])
train.train(train_data, test_data, model, loss_fn, optimizer, batch_size, num_epochs, msrs)
```

```python
for rec in msrs.get_train_recorder():
    if rec.get_name() == "Cross Entropy Sharpness":
        sharpness_02_v1 = rec.get_records()
#     if rec.get_name() == "Cross Entropy Sharpness V2":
#         sharpness_005_v2 = rec.get_records()
```

```python
for rec in msrs.get_train_recorder():
    if rec.get_name() == "Cross Entropy Sharpness":
        sharpness_04_v1 = rec.get_records()
#     if rec.get_name() == "Cross Entropy Sharpness V2":
#         sharpness_01_v2 = rec.get_records()
```

```python
from matplotlib import pyplot as plt

epc_idx = [y for y, x in sharpness_02_v1]

sval_02_v1 = np.array([x for y, x in sharpness_02_v1])
sval_04_v1 = np.array([x for y, x in sharpness_04_v1])
# sval_005_v2 = np.array([x for y, x in sharpness_005_v2])
# sval_01_v2 = np.array([x for y, x in sharpness_01_v2])

sval_02_v1 = np.minimum(20, sval_02_v1)
sval_04_v1 = np.minimum(20, sval_04_v1)
# sval_005_v2 = np.minimum(50, sval_005_v2)
# sval_01_v2 = np.minimum(50, sval_01_v2)

plt.scatter(epc_idx, sval_02_v1, label='eta=0.2_v1', color='yellow')
plt.scatter(epc_idx, sval_04_v1, label='eta=0.4_v1', color='red')

# plt.scatter(epc_idx, sval_005_v1, label='eta=0.05_v2', color='orange')
# plt.scatter(epc_idx, sval_01_v1, label='eta=0.1_v2', color='blue')
plt.axhline(y=10, color='yellow', linestyle='--')
plt.axhline(y=5, color='red', linestyle='--')
plt.legend()
plt.savefig('gd_eos.png')
```
